Import pandas and seaborn library


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import Military spending data
militaryspendingdata = pd.read_csv("militaryspending.csv") 

#Removing empty data
militaryspendingdata.dropna(inplace=True)
#Rename Column from YEAR 2012 to 2012
new_cols=['2012','2013','2014','2015','2016','2017']
militaryspendingdata.rename(columns=dict(zip(militaryspendingdata.columns[1:], new_cols)),inplace=True)
col_list = militaryspendingdata.columns[1:].values
#Replace $ sign in values
militaryspendingdata[col_list]=militaryspendingdata[col_list].replace({'\$': '', ',': ''}, regex=True)
#Convert Object type to numeric
militaryspendingdata[col_list]=militaryspendingdata[col_list].apply(pd.to_numeric)
#Plotting line graph
militaryspendingdata.plot(x=militaryspendingdata['Country'], y=col_list, kind='line',figsize=(12,4))
plt.title('Military Spending By Each Country Per Year')
plt.show()
```
